```python
import numpy as np
import imageio as imo
```

```python
sigma1 = 1.5
sigma2 = 5
sigma3 = 10
ro1=10
ro2=100

#calculating the filter size
msize1 = int(np.ceil(sigma1 * 2.575) * 2 + 1)
msize2 = int(np.ceil(sigma2 * 2.575) * 2 + 1)
msize3 = int(np.ceil(sigma3 * 2.575) * 2 + 1)
print(msize1)
print(msize2)
print(msize3)


#storing gaussian distribution in 2D array for later use
gauss2Ds1=np.zeros([msize1,msize1])
for i in range(0,msize1):
        for j in range(0,msize1):
            gauss2Ds1[i][j] = np.exp(-0.5 * (((i-msize1/2)**2+(j-msize1/2)**2)/(sigma1)**2))
            
gauss2Ds2=np.zeros([msize2,msize2])
for i in range(0,msize2):
        for j in range(0,msize2):
            gauss2Ds2[i][j] = np.exp(-0.5 * (((i-msize2/2)**2+(j-msize2/2)**2)/(sigma2)**2))
            
gauss2Ds3=np.zeros([msize3,msize3])
for i in range(0,msize3):
        for j in range(0,msize3):
            gauss2Ds3[i][j] = np.exp(-0.5 * (((i-msize3/2)**2+(j-msize3/2)**2)/(sigma3)**2))

# 2D gaussian with intensity factor
# (1/(np.sqrt(2*np.pi)*ro1))
# (1/(2*np.pi*sigma1*sigma1))

#performing filtering with Cat Img
f = imo.imread('cat.png')
#calculating number of rows and cols
rows = f.shape[0]
cols = f.shape[1]
print(rows)
print(cols)

#creating temporary place holder for the filterd images
#a particular tmpij means filtering done with ith sigma and jth ro
tmp11 = np.zeros_like(f)
tmp12 = np.zeros_like(f)

tmp21 = np.zeros_like(f)
tmp22 = np.zeros_like(f)

tmp31 = np.zeros_like(f)
tmp32 = np.zeros_like(f)

#looping for entire image 
for M in range(0,rows):
    for N in range(0,cols):
        #computations with sigma1 and ro1,ro2 
        #variables to hold sum and normalization factor
        sumx11 = 0
        norm11 = 0
        sumx12 = 0
        norm12 = 0
        #looping according to the filter size
        for i in range(-4,5):
            for j in range(-4,5):
                #handling the corner cases
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    #computing gaussian factor using Intensity difference as input
                    fact11 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    #multiply with the smoothing gaussian distribution
                    tmpfact11=fact11*gauss2Ds1[i+4][j+4]
                    sumx11+=tmpfact11*f[M+i][N+j]
                    norm11+=tmpfact11
                    
                    #same computation keeping the sigma same but changing the ro value
                    fact12 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact12=fact12*gauss2Ds1[i+4][j+4]
                    sumx12+=tmpfact12*f[M+i][N+j]
                    norm12+=tmpfact12
        #avoiding exception
        if norm11 != 0 :
            tmp11[M][N]=sumx11/norm11
        if norm12 != 0 :
            tmp12[M][N]=sumx12/norm12
            
        #computations with sigma2 and ro1,ro2 
        sumx21 = 0
        norm21 = 0
        sumx22 = 0
        norm22 = 0
        for i in range(-13,14):
            for j in range(-13,14):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact21 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact21=fact21*gauss2Ds2[i+13][j+13]
                    sumx21+=tmpfact21*f[M+i][N+j]
                    norm21+=tmpfact21
                    
                    fact22 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact22=fact22*gauss2Ds2[i+13][j+13]
                    sumx22+=tmpfact22*f[M+i][N+j]
                    norm22+=tmpfact22
        if norm21 != 0 :
            tmp21[M][N]=sumx21/norm21
        if norm22 != 0 :
            tmp22[M][N]=sumx22/norm22
                
        #computations with sigma3 and ro1,ro2 
        sumx31 = 0
        norm31 = 0
        sumx32 = 0
        norm32 = 0
        for i in range(-26,27):
            for j in range(-26,27):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact31 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact31=fact31*gauss2Ds3[i+26][j+26]
                    sumx31+=tmpfact31*f[M+i][N+j]
                    norm31+=tmpfact31

                    fact32 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact32=fact32*gauss2Ds3[i+26][j+26]
                    sumx32+=tmpfact32*f[M+i][N+j]
                    norm32+=tmpfact32
        if norm31 != 0 :
            tmp31[M][N]=sumx31/norm31
        if norm32 != 0 :
            tmp32[M][N]=sumx32/norm32

#Storing the filtered images
imo.imwrite('cats1ro1.png',tmp11)
imo.imwrite('cats1ro2.png',tmp12)
imo.imwrite('cats2ro1.png',tmp21)
imo.imwrite('cats2ro2.png',tmp22)
imo.imwrite('cats3ro1.png',tmp31)
imo.imwrite('cats3ro2.png',tmp32)


#With bauckhage Img
f = imo.imread('bauckhage.jpg')
rows = f.shape[0]
cols = f.shape[1]
print(rows)
print(cols)

tmp11 = np.zeros_like(f)
tmp21 = np.zeros_like(f)
tmp31 = np.zeros_like(f)

tmp12 = np.zeros_like(f)
tmp22 = np.zeros_like(f)
tmp32 = np.zeros_like(f)

for M in range(0,rows):
    for N in range(0,cols):
        sumx11 = 0
        norm11 = 0
        sumx12 = 0
        norm12 = 0
        for i in range(-4,5):
            for j in range(-4,5):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact11 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact11=fact11*gauss2Ds1[i+4][j+4]
                    sumx11+=tmpfact11*f[M+i][N+j]
                    norm11+=tmpfact11
                    
                    fact12 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact12=fact12*gauss2Ds1[i+4][j+4]
                    sumx12+=tmpfact12*f[M+i][N+j]
                    norm12+=tmpfact12
        if norm11 != 0 :
            tmp11[M][N]=sumx11/norm11
        if norm12 != 0 :
            tmp12[M][N]=sumx12/norm12
            
        
        sumx21 = 0
        norm21 = 0
        sumx22 = 0
        norm22 = 0
        for i in range(-13,14):
            for j in range(-13,14):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact21 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact21=fact21*gauss2Ds2[i+13][j+13]
                    sumx21+=tmpfact21*f[M+i][N+j]
                    norm21+=tmpfact21
                    
                    fact22 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact22=fact22*gauss2Ds2[i+13][j+13]
                    sumx22+=tmpfact22*f[M+i][N+j]
                    norm22+=tmpfact22
        if norm21 != 0 :
            tmp21[M][N]=sumx21/norm21
        if norm22 != 0 :
            tmp22[M][N]=sumx22/norm22
                
        
        sumx31 = 0
        norm31 = 0
        sumx32 = 0
        norm32 = 0
        for i in range(-26,27):
            for j in range(-26,27):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact31 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact31=fact31*gauss2Ds3[i+26][j+26]
                    sumx31+=tmpfact31*f[M+i][N+j]
                    norm31+=tmpfact31

                    fact32 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact32=fact32*gauss2Ds3[i+26][j+26]
                    sumx32+=tmpfact32*f[M+i][N+j]
                    norm32+=tmpfact32
        if norm31 != 0 :
            tmp31[M][N]=sumx31/norm31
        if norm32 != 0 :
            tmp32[M][N]=sumx32/norm32
            
imo.imwrite('bauckhages1ro1.jpg',tmp11)
imo.imwrite('bauckhages1ro2.jpg',tmp12)
imo.imwrite('bauckhages2ro1.jpg',tmp21)
imo.imwrite('bauckhages2ro2.jpg',tmp22)
imo.imwrite('bauckhages3ro1.jpg',tmp31)
imo.imwrite('bauckhages3ro2.jpg',tmp32)

#With Clock Img
f = imo.imread('clock.jpg')
rows = f.shape[0]
cols = f.shape[1]
print(rows)
print(cols)

tmp11 = np.zeros_like(f)
tmp21 = np.zeros_like(f)
tmp31 = np.zeros_like(f)

tmp12 = np.zeros_like(f)
tmp22 = np.zeros_like(f)
tmp32 = np.zeros_like(f)

for M in range(0,rows):
    for N in range(0,cols):
        sumx11 = 0
        norm11 = 0
        sumx12 = 0
        norm12 = 0
        for i in range(-4,5):
            for j in range(-4,5):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact11 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact11=fact11*gauss2Ds1[i+4][j+4]
                    sumx11+=tmpfact11*f[M+i][N+j]
                    norm11+=tmpfact11
                    
                    fact12 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact12=fact12*gauss2Ds1[i+4][j+4]
                    sumx12+=tmpfact12*f[M+i][N+j]
                    norm12+=tmpfact12
        if norm11 != 0 :
            tmp11[M][N]=sumx11/norm11
        if norm12 != 0 :
            tmp12[M][N]=sumx12/norm12
            
        
        sumx21 = 0
        norm21 = 0
        sumx22 = 0
        norm22 = 0
        for i in range(-13,14):
            for j in range(-13,14):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact21 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact21=fact21*gauss2Ds2[i+13][j+13]
                    sumx21+=tmpfact21*f[M+i][N+j]
                    norm21+=tmpfact21
                    
                    fact22 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact22=fact22*gauss2Ds2[i+13][j+13]
                    sumx22+=tmpfact22*f[M+i][N+j]
                    norm22+=tmpfact22
        if norm21 != 0 :
            tmp21[M][N]=sumx21/norm21
        if norm22 != 0 :
            tmp22[M][N]=sumx22/norm22
                
        
        sumx31 = 0
        norm31 = 0
        sumx32 = 0
        norm32 = 0
        for i in range(-26,27):
            for j in range(-26,27):
                if(M+i>=0 and M+i<rows and N+j>=0 and N+j<cols):
                    fact31 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro1)**2)
                    tmpfact31=fact31*gauss2Ds3[i+26][j+26]
                    sumx31+=tmpfact31*f[M+i][N+j]
                    norm31+=tmpfact31

                    fact32 = np.exp(-0.5 * (((int)(f[M][N])-(int)(f[M+i][N+j]))/ ro2)**2)
                    tmpfact32=fact32*gauss2Ds3[i+26][j+26]
                    sumx32+=tmpfact32*f[M+i][N+j]
                    norm32+=tmpfact32
        if norm31 != 0 :
            tmp31[M][N]=sumx31/norm31
        if norm32 != 0 :
            tmp32[M][N]=sumx32/norm32
            
imo.imwrite('clocks1ro1.jpg',tmp11)
imo.imwrite('clocks1ro2.jpg',tmp12)
imo.imwrite('clocks2ro1.jpg',tmp21)
imo.imwrite('clocks2ro2.jpg',tmp22)
imo.imwrite('clocks3ro1.jpg',tmp31)
imo.imwrite('clocks3ro2.jpg',tmp32)
```
